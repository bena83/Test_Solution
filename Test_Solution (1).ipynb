# Introduction

Whatever the size of your business, ensuring that it's successful or not is important to calculate the feasibility of its future, or just to give an insight into how effective it is and JustDice GmbH is no different. For example, some of the processes may need to be changed to make the business more profitable. Success can be measured in a number of different ways, and it's important to remember that what success means for one business, might not be the same for any other business.Success is often measured based on a number of realistic targets. These are often known as KPIs (key performance indicators). They are unique for each business because every business is different and will measure success in different ways.A company's success is often associated with its sales revenues and profits.Essentially for a company like JustDice GmbH to be profitable , the revenue must be more than the Expenses.From the data we will be looking at the total revenues against the expenses and use that to calculate the profits amongst other KPI's we will be looking at.

```python
sns.pairplot(df)
sns.set_style('dark')
plt.show()
```

```python
#A bar graph showing the total adspend per month in usd
```

```python
plt.figure(figsize=(15,10))
sns.set()
ax=sns.barplot(x=df['month'],y=df['value_usd'],ci=None,palette='CMRmap',estimator=np.sum)
plt.xticks(rotation=60)
plt.title('Total adspend per month\n',fontsize=20)
ax.bar_label(ax.containers[0],fmt='%.f')
plt.xlabel('month' ,size=16, family='monospace', weight=900)
plt.ylabel('value_usd' ,size=16, family='monospace',weight=900)
plt.xticks(rotation=45,weight=600)
plt.savefig('adspend_month.png')
plt.show()
```
